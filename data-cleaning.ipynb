```python
import pandas as pd
```

```python
_df = pd.read_csv("./data/raw/IS_production2.csv",
                  na_values=[-1]  # some of the rows have -1 to indicate missing values; if there are other such values you can add
                  )

_country = (_df.loc[:, "Economy"]
               .str
               .split('_', expand=True)
               .loc[:, 1])

hdf = (_df.rename({"Label2": "statistic"}, axis=1)
          .assign(country=_country)
          .drop(["Label1", "Economy", "Number"], axis=1)
          .set_index(["country", "statistic"])
          .sort_index()
          .stack(level=-1)
          .unstack(level=1))

(hdf.index
    .rename("year", level=1, inplace=True))

# add the additional features
_features = {"PerCapitaGDP": hdf.loc[:, "GDP"] / hdf.loc[:, "POP"],
             "PerCapitaITM": hdf.loc[:, "ITM"] / hdf.loc[:, "POP"]}
with_features_hdf = hdf.assign(**_features)
 
 
```
